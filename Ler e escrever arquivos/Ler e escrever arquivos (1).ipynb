# Escrevendo no Livro (write())
A função write() é usada para colocar texto dentro do arquivo.

```python
print("--- Escrevendo no Arquivo (Modo 'w') ---")
nome_arquivo_escrever = "meu_primeiro_arquivo.txt"
with open(nome_arquivo_escrever, "w", encoding="utf-8") as arquivo_escrita:
    arquivo_escrita.write("Olá, mundo dos arquivos \n")
    arquivo_escrita.write("Este é o meu primeiro arquivo criado com Python.\n")
    arquivo_escrita.write("Que legal aprender sobre isso!\n")
print(F"Texto escrito com sucesso no '{nome_arquivo_escrever}'.")

print("\n--- Anexando no Arquivo (Modo 'a') ---")

with open(nome_arquivo_escrever, 'a', encoding='utf-8') as arquivo_anexar:
    arquivo_anexar.write("Adicionando uma nova linha no final. \n")
    arquivo_anexar.write("Isso é o modo 'append' em ação")

print(f"Mais texto anexado com sucesso no '{nome_arquivo_escrever}'.")

```

Lendo o Livro (read(), readline(), readlines())
Quando você quer ler, tem algumas opções:

read(): Lê o conteúdo INTEIRO do arquivo como uma única string.

readline(): Lê apenas a próxima LINHA do arquivo. Útil quando o arquivo é muito grande.

readlines(): Lê TODAS as linhas do arquivo e retorna uma lista de strings, onde cada elemento da lista é uma linha do arquivo.

```python
print("\n --- Lendo o Arquivo (Modo 'r') ---")
nome_arquivo_ler = "meu_primeiro_arquivo.txt"
try:
     # 1. Lendo o conteúdo inteiro com read()
    print("\nUsando read() (lê tudo de uma vez):")
    with open(nome_arquivo_ler, 'r', encoding='utf-8') as arquivo_leitura:
        conteudo_completo = arquivo_leitura.read()
        print(conteudo_completo)
    #lendo linha por linha com readline()
    print("\nUsando readLine() (lê uma linha por vez):")
    with open(nome_arquivo_ler, 'r', encoding='utf-8') as arquivo_leitura:
        primeira_linha = arquivo_leitura.readline()
        segunda_linha = arquivo_leitura.readline()
        print(f"Primeira linha: {primeira_linha.strip()}")
        print(f"segunda linha: {segunda_linha.strip()}")

    # lendo todas as linhas em uma lista com readlines()
    print("\nUsando readlines()(lê todas as linhas em uma lista):")
    with open(nome_arquivo_ler, 'r', encoding='utf-8') as arquivo_leitura:
        lista_de_linhas = arquivo_leitura.readlines()
        for i, linha in enumerate(lista_de_linhas):
            print(f"Linha{i+1}: {linha.strip()}")
except FileNotFoundError:
    print(f"Erro: O arquivo '{nome_arquivo_ler}' não foi encontrado. Execute o exemplo de escrita primeiro!")
except Exception as e:
    print(f"Ocorreu um erro inesperado na leitura: {e}")
```

```python
import json
import os
print("---Desafio 1: Pessoas com json ---")
class Pessoa:
    def __init__(self,nome,  idade):
        self.nome = nome
        self.idade = idade
        def __repr__(self):
            return f"Pessoa(nome='{self.nome}', idade={self.idade})"
    @staticmethod
    def salvar_para_json(lista_pessoas, nome_arquivo):
        lista_de_dicionarios = []
        for pessoa in lista_pessoas:
            lista_de_dicionarios.append({"nome": pessoa.nome, "idade": pessoa.idade})
            with open(nome_arquivo, 'w', encoding='utf-8') as f:
                json.dump(lista_de_dicionarios, f, indent=4, ensure_ascii=False)
            print(f"Pessoas salvas em '{nome_arquivo}'.")
    @staticmethod
    def carregar_de_json(nome_arquivo):
        lista_pessoas =[]
        with open(nome_arquivo, 'r', encoding='utf-8') as f:
            lista_de_dicionarios = json.load(f)
            for dados_pessoa in lista_de_dicionarios:
                lista_pessoas.append(Pessoa(dados_pessoa['nome'], dados_pessoa['idade']))
            return lista_pessoas
# 1. Criar instâncias de Pessoa
pessoas_originais = [
    Pessoa("Alice", 30),
    Pessoa("Bruno", 25),
    Pessoa("Carla", 28),
    Pessoa("Daniel", 35)
]

# 2. Salvar para JSON
nome_arquivo_json = "minhas_pessoas.json"
Pessoa.salvar_para_json(pessoas_originais, nome_arquivo_json)

# 3. Carregar de JSON
pessoas_carregadas = Pessoa.carregar_de_json(nome_arquivo_json)

# 4. Verificar
print("\nPessoas carregadas do JSON:")
for p in pessoas_carregadas:
    print(f"- Nome: {p.nome}, Idade: {p.idade}")

print("-" * 40)   
    
              
      
    
      
```

# XML

```python
import xml.etree.ElementTree as ET
import os #


print("\n--- Desafio 2: Carros com XML (Corrigido) ---")

class Carro:
    def __init__(self, nome, potencia):
        self.nome = nome
        self.potencia = potencia

    def __repr__(self):
        return f"Carro(nome='{self.nome}', potencia={self.potencia})"

    @staticmethod
    def salva_carros_xml(caminho_arquivo, *carros_a_salvar):
    
        raiz_carros = ET.Element("MinhaGaragemDeCarros") 
        
       
        for carro in carros_a_salvar:
            no_carro = ET.SubElement(raiz_carros, 'Carro')
            
            ET.SubElement(no_carro, 'Nome').text = carro.nome
            ET.SubElement(no_carro, 'Potencia').text = str(carro.potencia) # Potência é int, mas XML armazena texto

        arvore_garagem = ET.ElementTree(raiz_carros)
        
        with open(caminho_arquivo, 'wb') as files:
            arvore_garagem.write(files, encoding='utf-8', xml_declaration=True)
        print(f"Carros salvos em '{caminho_arquivo}'.")

    @staticmethod
    def le_carros_xml(caminho_arquivo):
      
        lista_carros_lidos = []
        tree = ET.parse(caminho_arquivo)
        root = tree.getroot()
        
        for carro_element in root.findall('Carro'):
            nome_lido = None
            potencia_lida = None
            for dado in carro_element:
                if dado.tag == 'Nome':
                    nome_lido = dado.text 
                elif dado.tag == 'Potencia':
                    potencia_lida = int(dado.text) 
            if nome_lido and potencia_lida is not None:
                lista_carros_lidos.append(Carro(nome_lido, potencia_lida))
        return lista_carros_lidos

# --- Execução do Desafio ---

# 1. Criar instâncias de Carro (objetos Python em memória)
carros_originais = [
    Carro("Civic", 150),
    Carro("Corolla", 140),
    Carro("HB20", 120)
]
print(f"Carros originais (objetos Python): {carros_originais}")

# 2. Salvar para XML (serializaão> objetos Python ---> XML no arquivo)
nome_arquivo_xml = "meus_carros.xml"
Carro.salva_carros_xml(nome_arquivo_xml, *carros_originais)
# 3. Carregar de XML (desserialização: XML no arquivo -> objetos Python)
carros_carregados = Carro.le_carros_xml(nome_arquivo_xml)

  # 4. Verificar (mostra os objetos Python reconstruídos)
print("\nCarros carregados do XML (objetos Python reconstruídos):")
for c in carros_carregados:
    print(f"- Nome: {c.nome}, Potência: {c.potencia} HP")

print("-" * 40)
```

# YAML

```python
import yaml
import os
print("\n --- Desafio 3: Configurações de Servidor com YAML (comentado) ---")
config_servidor_original = {
    'servidor':{
    'nome': 'servidor_producao_01',
    'ip': '192.168.1.100',
    'porta': 8080,
    'servicos_ativos': ['web', 'database', 'cache']
    },
    'banco_de_dados': {
        'tipo': 'PostgreSQL',
        'usuario': 'admin_db',
        'senha': 'senhaSegura123',
        'limite_conexoes': 100
    },
    'monitoramento': True,
    'notificacoes': {
        'email': ['dev@empresa.com', 'ops@empresa.com'],
        'sms_habilitado': False
    }
}

print("Dicionário Python de configurações criado.")
# 2. Salvar para YAML (Serialização: Dicionário Python -> YAML no arquivo)
# 2. Salvar para YAML (Serialização: Dicionário Python -> YAML no arquivo)
#    yaml.dump() converte o dicionário Python em uma string YAML e a escreve no arquivo.
#    'default_flow_style=False': Garante que o YAML seja salvo no formato de bloco (indentado),
#                                o que é mais legível para configurações.
#    'allow_unicode=True': Garante que caracteres especiais (acentos, emojis, etc.) sejam
#                          salvos corretamente no arquivo.
nome_arquivo_yaml_config = "config_servidor.yaml"
with open(nome_arquivo_yaml_config, 'w', encoding='utf-8') as f:
    yaml.dump(config_servidor_original, f, default_flow_style=False, allow_unicode=True)
print(f"configurações de servidor salvas em 'nome_arquivo_yaml_config'.")

# 3. Carregar de YAML (Desserialização: YAML no arquivo -> Dicionário Python)
#    yaml.safe_load() lê o conteúdo do arquivo YAML e o converte de volta em um dicionário Python.
#    Usar 'safe_load' é recomendado para segurança, pois impede a execução de código arbitrário.

with open(nome_arquivo_yaml_config, 'r', encoding='utf-8') as f:
    config_servidor_carregada = yaml.safe_load(f)
print(f"Configurações de servidor carregadas de '{nome_arquivo_yaml_config}'.")

# 4. Acessar e imprimir informações (verificação do dicionário Python carregado)
#    Agora que o YAML foi desserializado para um dicionário Python,
#    podemos acessar seus dados usando a sintaxe de dicionário normal (chaves).

print("\n Informações carregadas do YAML (acessando o dicionário python);")
# 4. Acessar e imprimir informações (verificação do dicionário Python carregado)
#    Agora que o YAML foi desserializado para um dicionário Python,
#    podemos acessar seus dados usando a sintaxe de dicionário normal (chaves).
print("\nInformações carregadas do YAML (acessando o dicionário Python):")
print(f"- Nome do Servidor: {config_servidor_carregada['servidor']['nome']}")
print(f"- Porta do Servidor: {config_servidor_carregada['servidor']['porta']}")
print(f"- Tipo de Banco de Dados: {config_servidor_carregada['banco_de_dados']['tipo']}")
print(f"- Primeiro Serviço Ativo: {config_servidor_carregada['servidor']['servicos_ativos'][0]}")
print(f"- Primeiro E-mail de Notificação: {config_servidor_carregada['notificacoes']['email'][0]}")


print("-" * 40)



    
    
        

```
